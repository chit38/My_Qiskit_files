```python
from qiskit import *
```

```python
%config InlineBackend.figure_format = 'svg' # Makes the images look nice
```

```python
qc = QuantumCircuit()
qa = QuantumRegister(2,'a')
qb = QuantumRegister(1,'b')
```

```python
qc.add_register(qa)
qc.add_register(qb)
```

```python
qc.qregs
```

```python
qc.draw(output = 'mpl')
```
